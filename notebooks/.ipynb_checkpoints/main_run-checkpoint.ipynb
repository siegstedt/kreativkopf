```python
# prepare keyword dict
def prepare_keywords():
    # Construct selector object and retrieve traffic estimates.
    keywords = [
      {'text': 'Landmaschine', 'matchType': 'BROAD'},
      {'text': 'Holzspalter', 'matchType': 'PHRASE'},
      {'text': 'Mulcher', 'matchType': 'EXACT'},
      {'text': 'Mulcher', 'matchType': 'BROAD'},
      {'text': 'Mulcher', 'matchType': 'PHRASE'},
    ]
    negative_keywords = [
      {'text': 'Traktor', 'matchType': 'BROAD'}
    ]
    keyword_estimate_requests = []
    for keyword in keywords:
    keyword_estimate_requests.append({
        'keyword': {
            'xsi_type': 'Keyword',
            'matchType': keyword['matchType'],
            'text': keyword['text']
        }
    })

    for keyword in negative_keywords:
    keyword_estimate_requests.append({
        'keyword': {
            'xsi_type': 'Keyword',
            'matchType': keyword['matchType'],
            'text': keyword['text']
        },
        'isNegative': 'true'
    })
```

```python
# load estimate
def load_campaign_estimate(client, keyword_estimate_requests, locations):
    # Initialize appropriate service.
    traffic_estimator_service = client.GetService(
      'TrafficEstimatorService', version='v201809')

    # Create ad group estimate requests.
    adgroup_estimate_requests = [{
      'keywordEstimateRequests': keyword_estimate_requests,
      'maxCpc': {
          'xsi_type': 'Money',
          'microAmount': '1000000' # which is equal to 1 EUR
      }
    }]

    # Create campaign estimate requests.
    campaign_estimate_requests = [{
      'adGroupEstimateRequests': adgroup_estimate_requests,
      'criteria': [
          {
              'xsi_type': 'Location',
              'id': '1004101'  # Aschaffenburg.
          },
          {
              'xsi_type': 'Language',
              'id': '1001'  # German.
          }
      ],
    }]

    # Create the selector.
    selector = {
      'campaignEstimateRequests': campaign_estimate_requests,
    }

    # Optional: Request a list of campaign-level estimates segmented by
    # platform.
    selector['platformEstimateRequested'] = True

    # Get traffic estimates.
    estimates = traffic_estimator_service.get(selector)
```
