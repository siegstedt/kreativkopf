```python
from pytrends.request import TrendReq
import pandas as pd
import time
```

```python
pytrend = TrendReq(hl='de-DE', tz=360)

dataset = []
pytrend.build_payload(
    kw_list=['garten'],
    cat=0,
    timeframe='2020-04-01 2020-04-10',
    geo='DE'
)
data = pytrend.interest_over_time()
if not data.empty:
    data = data.drop(labels=['isPartial'],axis='columns')
    dataset.append(data)

pd.concat(dataset, axis=1)
```

```python
d = pytrend.related_queries()
pd.concat([
    pd.DataFrame(d['garten']['top']),
    pd.DataFrame(d['garten']['rising']),
], axis=1
)
```

```python
d = pytrend.related_topics()
pd.concat([
    pd.DataFrame(d['garten']['top']),
    pd.DataFrame(d['garten']['rising']),
], axis=1
)
```
