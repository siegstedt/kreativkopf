```python
from pytrends.request import TrendReq
import pandas as pd
```

```python
keyword_list = ['garten', 'gartenbau', 'gartengestaltung', 'leistungen', 'kontakt']
```

```python
def _parse_related_queries_response(keyword_dict, item, value, number_of_results):
    if keyword_dict[item] is None:
        return None
    else: 
        return [
            i
            for i
            in keyword_dict[item][value][:number_of_results] 
        ]

    
def generate_related_queries_from_keywords(keyword_list, number_of_results=3):
    
    pytrend = TrendReq(hl='de-DE')

    data = []
    for keyword in keyword_list:
        pytrend.build_payload(
            kw_list = [keyword],
            timeframe = 'today 1-m',
            geo='DE'
        )
        data.append(pytrend.related_queries())
    
    out_data = [
        {
            keyword: {
                'top_query': _parse_related_queries_response(data[index][keyword], 'top', 'query', number_of_results),
                'top_value': _parse_related_queries_response(data[index][keyword], 'top', 'value', number_of_results),
                'rising_query': _parse_related_queries_response(data[index][keyword], 'rising', 'query', number_of_results),
                'rising_value': _parse_related_queries_response(data[index][keyword], 'rising', 'value', number_of_results),
            }
        }
        for index, keyword in 
        enumerate(keyword_list)    
    ]

    return out_data
```

```python
generate_related_queries_from_keywords(keyword_list=keyword_list, number_of_results=5)
```
