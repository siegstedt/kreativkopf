#### channel conversion

```python
import pandas as pd
conversion_map = {
    'general': {
        'channel': ['facebook', 'google', 'instagram'],
        'impression_to_visit': [0.1, 0.1, 0.1],
        'visit_to_cart': [0.1, 0.1, 0.1],
        'cart_to_purchase': [0.1, 0.1, 0.1],
    },
    'haus- und garten': {
        'channel': ['facebook', 'google', 'instagram'],
        'impression_to_visit': [0.2, 0.2, 0.2],
        'visit_to_cart': [0.2, 0.2, 0.2],
        'cart_to_purchase': [0.01, 0.01, 0.01],
    },

}
```

```python
conversions = pd.DataFrame()
for i in conversion_map:
    #df = df.append(pd.DataFrame(conversion_map[i])).reset_index(drop=True)
    df = pd.DataFrame(conversion_map[i])
    df['scope'] = i
    df = df.set_index(['scope','channel'])
    conversions = conversions.append(df)
conversions
```
