```python
df[['daily_cost','daily_clicks']].plot(kind="scatter",x='daily_cost',y='daily_clicks',figsize=(15,10))
```

```python
df['cum_cost'] = df['daily_cost'].cumsum()
df['on_cost'] = [
    True if row < 1000 else False
    for row
    in df['cum_cost']
] 
df
# besser auf Kosten optimieren
# ODER auf reach optimieren
```

```python
g_keywords = list(df[df['on_cost'] == True]['keyword'].unique())
g_reach_estimate = df[df['on_cost'] == True]['reach'].sum()
g_click_through = df[df['on_cost'] == True]['click_through'].mean()
g_est_clicks = df[df['on_cost'] == True]['daily_clicks'].sum()
```

## budget split

```python
pd.options.mode.chained_assignment = None
```

```python
estimates_df = pd.DataFrame()
estimates_data = fb_estimate + g_estimate
for i in estimates_data:
    estimates_df = estimates_df.append(pd.DataFrame(i).transpose())
estimates_df
```

```python
# generate conversion data
conversion_map = {
    'general': {
        'channel': ['facebook', 'google', 'instagram'],
        'visit_to_cart': [0.1, 0.1, 0.1],
        'cart_to_purchase': [0.1, 0.1, 0.1],
    },
    'haus- und garten': {
        'channel': ['facebook', 'google', 'instagram'],
        'visit_to_cart': [0.2, 0.1, 0.2],
        'cart_to_purchase': [0.5, 0.2, 0.01],
    },
}

# megre the converions into estimates
conversions = pd.DataFrame()
for i in conversion_map:
    df = pd.DataFrame(conversion_map[i])
    df['scope'] = i
    df = df.set_index(['scope','channel'])
    conversions = conversions.append(df)
conversions_df = conversions.loc['haus- und garten',:].reset_index().set_index('channel',drop=True)
estimates_df = estimates_df.merge(conversions_df, how='left', left_on=estimates_df.index, right_on=conversions_df.index).rename(columns={'key_0':'channel'})

# prepare conversion data
estimates_df['est_cart_items'] = estimates_df['est_click'] * estimates_df['visit_to_cart']
estimates_df['est_purchased_items'] = estimates_df['est_cart_items'] * estimates_df['cart_to_purchase']
# clean up
estimates_df = estimates_df[[
    'channel', 'keywords', 'location', 'reach_estimate',
    'click_through','est_click', 'visit_to_cart','est_cart_items',
    'cart_to_purchase', 'est_purchased_items'
]]
```

```python
# budget split
estimates_df['budget_split'] = estimates_df['est_purchased_items'] / estimates_df['est_purchased_items'].sum()
```
